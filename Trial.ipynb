```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

def plot_response(fs, w, h, title):
    "Utility function to plot response functions"
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(0.5*fs*w/np.pi, 20*np.log10(np.abs(h)))
    ax.set_ylim(-40, 5)
    ax.set_xlim(0, 0.5*fs)
    ax.grid(True)
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('Gain (dB)')
    ax.set_title(title)


fs = 22050.0       # Sample rate, Hz
##Frequency for cut-off
cutoff = 7000.0    # Desired cutoff frequency, Hz
trans_width = 100  # Width of transition from pass band to stop band, Hz
##numtaps if low will give plots with troughs, need tobe fairly high for a good response
numtaps = 500      # Size of the FIR filter.
taps = signal.remez(numtaps, [0, cutoff, cutoff + trans_width, 0.5*fs], [1, 0], Hz=fs)
w, h = signal.freqz(taps, [1], worN=2000)
plot_response(fs, w, h, "Low-pass Filter")
plt.show()


time = np.arange(0, 20, 0.2)
sig = np.cos(time)
plt.plot(time, sigFilter)

plt.plot(time, sig, '-k', time, sigFilter, '-r')
```
